# Paper draft figure gallery

Loads all PNGs and MP4s in this directory and displays them with captions from
`figure_captions.yaml`. Use this to review the latest paper-ready figures in one place.

```python
from pathlib import Path
import yaml
from IPython.display import Image, Video, display, HTML

# Run from repo root or from output/gallery
_g = Path.cwd() / "output" / "gallery"
GALLERY_DIR = _g if (_g / "figure_captions.yaml").exists() else Path.cwd()
CAPTIONS_FILE = GALLERY_DIR / "figure_captions.yaml"

with open(CAPTIONS_FILE) as f:
    CAPTIONS = yaml.safe_load(f) or {}

def caption_for(name: str) -> str:
    return (CAPTIONS.get(name) or CAPTIONS.get(name.replace(" ", "_")) or "").strip() or "(no caption)"
```

```python
pngs = sorted(GALLERY_DIR.glob("*.png"))
mp4s = sorted(GALLERY_DIR.glob("*.mp4"))
print(f"PNGs: {len(pngs)}", pngs)
print(f"MP4s: {len(mp4s)}", mp4s)
```

## PNG figures

```python
for p in pngs:
    name = p.name
    cap = caption_for(name)
    display(HTML(f"<p><b>{name}</b></p>"))
    display(Image(filename=str(p)))
    display(HTML(f"<p><em>{cap}</em></p><hr/>"))
```

## MP4 videos

```python
for v in mp4s:
    name = v.name
    cap = caption_for(name)
    display(HTML(f"<p><b>{name}</b></p>"))
    display(Video(str(v), width=700))
    display(HTML(f"<p><em>{cap}</em></p><hr/>"))
```
