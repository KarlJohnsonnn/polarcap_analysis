# 05-00 - Shared Data Loading

This notebook centralizes the data loading and configuration parsing for the process budget analysis (Views A-E). It sets up the environment, loads the YAML configuration, discovers and opens the dataset, and builds the `rates_by_exp` dictionary containing all process rates.

Run this notebook first, or import its logic, before running any of the specific view notebooks (05a-05e).

```python
from __future__ import annotations
import sys
from pathlib import Path

REPO_ROOT = Path.cwd()
while REPO_ROOT != REPO_ROOT.parent and not (REPO_ROOT / '.git').exists():
    REPO_ROOT = REPO_ROOT.parent
SRC_DIR = REPO_ROOT / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

from utilities.process_budget_data import load_process_budget_data, stn_label, select_rates_for_range
import numpy as np
from dask.diagnostics import ProgressBar

CONFIG_YAML = REPO_ROOT / "notebooks" / "config" / "process_budget.yaml"
cfg = load_process_budget_data(REPO_ROOT, config_path=CONFIG_YAML)

# Unpack from cfg (all keys lowercase). View notebooks (05a–05e) use these names.
with ProgressBar():
    ds = cfg.get("ds").compute()
rates_by_exp = cfg.get("rates_by_exp", {})
size_ranges = cfg.get("size_ranges", {})
plot_range_keys = cfg.get("plot_range_keys", ["ALLBB"])
active_range_key = cfg.get("active_range_key", "ALLBB")
plot_exp_ids = cfg.get("plot_exp_ids", [1])
plot_stn_ids = cfg.get("plot_stn_ids", [0, 1, 2])
exp_idx = cfg.get("exp_idx", 1)
stn_idx = cfg.get("stn_idx", 0)
seed_start = cfg.get("seed_start")
time_coarsen = cfg.get("time_coarsen", "30s")
rate_floor = float(cfg.get("rate_floor", 1e-18))
station_labels = cfg.get("station_labels", {0: "S1", 1: "S2", 2: "S3"})
experiment_meta = cfg.get("experiment_meta", [])
diameter_um = cfg.get("diameter_um")
time_window = cfg.get("time_window", [])
height_sel_m = cfg.get("height_sel_m", [1400, 1200, 1000, 800])

print("Shared data loaded successfully.")
```
