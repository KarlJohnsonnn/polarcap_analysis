# 05 - Process Budget (Index)

**Scientific question (article storyline):** Which processes contribute the most to **ice crystal number growth**, **ice crystal mass growth**, and—for liquid—**droplet number and mass reduction**?

This notebook has been split into several smaller notebooks, each focusing on a specific view of the process budget:

* **[05-00-shared-data.ipynb](05-00-shared-data.ipynb)**: Shared data loading and configuration.
* **[05a-stacked-area.ipynb](05a-stacked-area.ipynb)**: View A – Normalised stacked area (fractional budget over time).
* **[05b-stacked-bars.ipynb](05b-stacked-bars.ipynb)**: View B – Stacked bar charts (budget by height).
* **[05c-dominance-confidence.ipynb](05c-dominance-confidence.ipynb)**: View C – Dominant process + confidence (time × height).
* **[05d-spectral-waterfall.ipynb](05d-spectral-waterfall.ipynb)**: View D – Bin-resolved spectral budget waterfall.
* **[05e-hovmoller.ipynb](05e-hovmoller.ipynb)**: View E – Hovmöller total-rate heatmap with dominant-process overlay.
* **[05-heat-release.ipynb](05-heat-release.ipynb)**: Heat release contour (time × altitude).

Run `05-00-shared-data.ipynb` first, or run the individual view notebooks directly (they import the shared data loading logic).
