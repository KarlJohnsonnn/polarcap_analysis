```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Wedge, Patch

```

```python
def draw_pie(ax, ratios, x, y, size):
    start_angle = 0
    for ratio, color in zip(ratios, ['blue', 'green', 'orange']):
        end_angle = start_angle + ratio * 360
        wedge = Wedge(center=(x, y), r=size, theta1=start_angle, theta2=end_angle, color=color, ec='black')
        ax.add_patch(wedge)
        start_angle = end_angle
```

```python


# Example data
time = np.linspace(0, 60, 10)
altitude = np.linspace(1, 15, 7)
data = np.random.dirichlet([1, 1, 1], size=(len(time), len(altitude)))

# Load the COSMO-SPECS data 
```

```python

fig, ax = plt.subplots(figsize=(10, 4))
ax.set_xlabel('Time (min)')
ax.set_ylabel('Altitude (km)')

# Add pie charts
for i, t in enumerate(time):
    for j, alt in enumerate(altitude):
        draw_pie(ax, data[i, j], t, alt, size=0.6)

# Add legend for pie chart processes
legend_elements = [Patch(facecolor='blue', label='Process 1'),
                   Patch(facecolor='green', label='Process 2'),
                   Patch(facecolor='orange', label='Process 3')]
ax.legend(handles=legend_elements, loc='upper right', title='Processes')

# Add scatter for flow rates
ax.scatter([], [], s=200, c='gray', label='8000 kg s⁻¹ m⁻¹')
ax.scatter([], [], s=100, c='gray', label='4000 kg s⁻¹ m⁻¹')

# Finalize plot
ax.set_xlim(0, 60)
ax.set_ylim(0, 16)
ax.grid(True)
plt.tight_layout()
plt.show()
```
